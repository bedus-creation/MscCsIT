### Import necessary files and library

```python
import numpy as np
import matplotlib.pyplot as plt
from Perceptron import Perceptron
from TrollPerceptron import TrollPerceptron
```

### Troll

```python
training_inputs = []
training_inputs.append(np.array([0]))
training_inputs.append(np.array([1]))
training_inputs.append(np.array([2]))
training_inputs.append(np.array([3]))


labels = np.array([2, 4,6,8])

perceptron = TrollPerceptron(1,1000)
perceptron.train(training_inputs, labels)

inputs = np.array([5])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))
```

```python
error = perceptron.getSquareError()
plt.plot(error)
```

### Training datasets for AND gate

```python
training_inputs = []
training_inputs.append(np.array([1, 1]))
training_inputs.append(np.array([1, 0]))
training_inputs.append(np.array([0, 1]))
training_inputs.append(np.array([0, 0]))

labels = np.array([1, 0, 0, 0])
```

### Train perceptron

```python
perceptron = Perceptron(2,10)
perceptron.train(training_inputs, labels)
```

### Let's test the preceptron

```python
inputs = np.array([0, 0])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

inputs = np.array([0, 1])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

inputs = np.array([1, 0])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

inputs = np.array([1, 1])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))
```

```python
squareError= perceptron.getSquareError();
plt.plot(squareError)
```

```python
weights = perceptron.getWeights()
print(weights)
```

### Let's try OR Gate

```python
training_inputs = []
training_inputs.append(np.array([1, 1]))
training_inputs.append(np.array([1, 0]))
training_inputs.append(np.array([0, 1]))
training_inputs.append(np.array([0, 0]))

labels = np.array([1, 1, 1, 0])
```

### OR gate test

```python
perceptron = Perceptron(2)
perceptron.train(training_inputs, labels)

inputs = np.array([0, 0])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

inputs = np.array([0, 1])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

inputs = np.array([1, 0])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

inputs = np.array([1, 1])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))
```

```python
error= perceptron.getErrors();
plt.plot(error)
```

```python
# XOR Gate
training_inputs = []
training_inputs.append(np.array([1, 1]))
training_inputs.append(np.array([1, 0]))
training_inputs.append(np.array([0, 1]))
training_inputs.append(np.array([0, 0]))

labels = np.array([1, 0, 0, 1])

perceptron = Perceptron(2,100,0.1)
perceptron.train(training_inputs, labels)


inputs = np.array([0, 0])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

inputs = np.array([0, 1])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

inputs = np.array([1, 0])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

inputs = np.array([1, 1])
print("inputs "+str(inputs)+" Output "+str(perceptron.predict(inputs)))

squareError= perceptron.getSquareError();
plt.plot(squareError)
```
